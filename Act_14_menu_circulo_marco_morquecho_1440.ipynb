* El presente problema consiste en desarrollar un programa interactivo que permita calcular el área de un círculo, su perímetro (circunferencia) y el volumen de una esfera. Para lograrlo, se emplea un enfoque procedural basado en funciones modulares, lo que facilita la organización del código y su reutilización. Cada cálculo se encapsula en una función independiente que solicita el radio únicamente cuando es invocada, manteniendo el principio de responsabilidad única.

* La estructura del programa gira en torno a una función menu() que se ejecuta dentro de un bucle while True, garantizando que la aplicación permanezca activa hasta que el usuario decida salir. Dentro de este bucle, se utilizan estructuras condicionales if, elif y else para dirigir el flujo de ejecución hacia la función correspondiente según la opción seleccionada. Este diseño es ideal para principiantes, ya que refuerza conceptos clave como el control de flujo, la llamada a funciones y la interacción básica con el usuario mediante input() y print().
* Al integrar emojis y un diseño limpio en la consola, se busca hacer la experiencia más amigable y motivadora, cumpliendo con el objetivo de crear una herramienta educativa elegante, funcional y alineada con las competencias digitales esperadas en educación media superior.


```python
# ==========================
# PROGRAMACION DE DE CALCULOS GEOMETRICOS
# elaborado por marco morquecho
# formulas area del circulo, perimetro del circulo, volumen de la esfera


import math



def areacirculo():
    """Calcula y muestra el área de un círculo."""
    radio = float(input("\n📏 Ingresa el radio del círculo: "))
    area = math.pi * radio ** 2
    print(f"🟢 El área del círculo es: {area:.2f} unidades²")

def perimetrocirculo():
    """Calcula y muestra el perímetro de un círculo."""
    radio = float(input("\n📏 Ingresa el radio del círculo: "))
    perimetro = 2 * math.pi * radio
    print(f"🔵 El perímetro del círculo es: {perimetro:.2f} unidades")

def volumenesfera():
    """Calcula y muestra el volumen de una esfera."""
    radio = float(input("\n📏 Ingresa el radio de la esfera: "))
    volumen = (4 / 3) * math.pi * radio ** 3
    print(f"🌐 El volumen de la esfera es: {volumen:.2f} unidades³")

def menu():
    """Muestra el menú interactivo y controla el flujo del programa."""
    while True:
        print("\n" + "="*40)
        print("🧮  MENÚ DE CÁLCULO GEOMÉTRICO  🧮")
        print("="*40)
        print("1️⃣  Calcular área del círculo")
        print("2️⃣  Calcular perímetro del círculo")
        print("3️⃣  Calcular volumen de la esfera")
        print("4️⃣  Salir")

        opcion = int(input("👉 Selecciona una opción (1-4): "))

        if opcion == 1:
            areacirculo()
        elif opcion == 2:
            perimetrocirculo()
        elif opcion == 3:
            volumenesfera()
        elif opcion == 4:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. 🌟")
            break
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un número del 1 al 4.")
            print("programa realizado por marco morquecho")

# Ejecución principal
if __name__ == "__main__":
    menu()
```
